# project summary 
The Household Consumption Expenditure Survey(HCES) in India provides insights into household spending patterns and living standards. Recent data from the 2023-24 survey indicates an increase in Monthly Per Capita Expenditure (MPCE) in both rural and urban areas, with a decrease in the urban-rural gap. The survey also reveals a decline in consumption inequality in both rural and urban areas.



# Problem statement :
The HCES data (Household Consumption Expenditure Survey) faces several problem statements including underreporting of certain expenditures, non-sampling errors, outdated definitions, and potential for manipulation or suppression for political reasons. Furthermore, discrepancies between NSSO and National Accounts data, and challenges in comparing surveys with different recall periods or definitions (rural areas, poverty lines) also pose issues, according to Shankar IAS Parliament.

```python
def outliers(data):
  outlier_col={}
  col=data.select_dtypes(include=[int,float]).columns
  for i in col:
    mean=data[i].mean()
    std=data[i].std()
    z_score=(data[i]-mean)/std
    outliers=data[np.abs(z_score) >3]
    if not outliers.empty:
      outlier_col[i]=outliers.index.tolist()
  return outlier_col

data=df
print(outliers(data))
wvt-isas-xmb
```
