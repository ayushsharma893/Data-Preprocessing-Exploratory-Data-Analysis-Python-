```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# find out the outlier 
col_num=df.select_dtypes(include=['int','float'])
for  column in col_num.columns:
    sns.boxplot(col_num[column])

    
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df1 = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(12, 8))
sns.heatmap(df1.corr(), annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.histplot(df['price'], kde=True)
plt.title('Price Distribution')
plt.show()
```

```python
top10_hp = df.sort_values(by='horsepower', ascending=False).head(70)

sns.scatterplot(x='horsepower', y='price', data=top10_hp)
plt.title('Price vs Horsepower')
plt.show()

sns.scatterplot(x='engine-size', y='price', data=df)
plt.title('Price vs Engine Size')
plt.show()
```

```python
sns.boxplot(x='body-style', y='price', data=df)
plt.title('Price by Body Style')

plt.show()

sns.boxplot(x='fuel-type', y='price', data=df)
plt.title('Price by Fuel Type')
plt.show()
```

```python
avg_price = df.groupby('make')['price'].mean().sort_values(ascending=False)
avg_price.plot( figsize=(12, 6))
plt.title('Average Car Price by Make')
plt.ylabel('Average Price')
plt.show()
```

```python
avg_price=df.groupby(['body-style','length','num-of-doors','engine-type'])['price'].mean().sort_values(ascending=False)
avg_price.plot(figsize=(17,8))
plt.title('Average Car Price by body-style ')
plt.ylabel('Average Price')
plt.show()
```
